# 📊 CRM Lead Conversion Analysis (Python EDA)

## 📌 Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/crm_leads.csv')
df.head()
```

## 📌 Data Check

```python
df.info()
df.describe()
```

```python
print(df.isnull().sum())
```

```python
print("Duplicates:", df.duplicated().sum())
```

## 📊 Lead Score Distribution

```python
plt.hist(df['LeadScore'])
plt.title("Lead Score Distribution")
plt.xlabel("Lead Score")
plt.ylabel("Frequency")
plt.show()
```

## 📊 Conversion Rate by Source

```python
conversion_by_source = df.groupby('Source')['Converted'].mean()

conversion_by_source.plot(kind='bar')
plt.title("Conversion Rate by Source")
plt.ylabel("Conversion Rate")
plt.show()
```

## 📊 Response Time Analysis

```python
df.groupby('Converted')['ResponseTimeHours'].mean().plot(kind='bar')
plt.title("Average Response Time (Converted vs Not)")
plt.ylabel("Hours")
plt.show()
```

## 📊 Revenue Distribution

```python
converted_df = df[df['Converted'] == 1]

plt.hist(converted_df['DealValue'])
plt.title("Revenue Distribution (Converted Leads)")
plt.xlabel("Deal Value")
plt.ylabel("Frequency")
plt.show()
```

## 📊 Industry-wise Conversion

```python
df.groupby('Industry')['Converted'].mean().plot(kind='bar')
plt.title("Conversion Rate by Industry")
plt.ylabel("Conversion Rate")
plt.show()
```

## 📌 Key Insights

- Higher lead scores show better conversion probability  
- Faster response time improves conversion outcomes  
- Referral and LinkedIn sources perform better
